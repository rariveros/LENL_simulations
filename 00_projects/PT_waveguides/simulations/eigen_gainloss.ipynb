```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
gamma = 10.0
k = 1.0
betas = np.arange(-0.9, 0.9, 0.02)
N = 15.74
for i in range(len(betas)):
    beta = betas[i]
    a = 2 * beta
    c = 2 * N * np.sqrt(k ** 2 - beta ** 2)
    d = 4 * gamma / (N + 2) ** 2  - 2 * np.sqrt(k ** 2 - beta ** 2) / N
    plt.scatter(beta, np.real(np.sqrt(c * d - 2 * a ** 2 + 0j)), color="b")
    plt.scatter(beta, np.imag(np.sqrt(c * d - 2 * a ** 2 + 0j)), color="r")
    plt.scatter(beta, np.real(np.sqrt(c * d + 0j)), color="b", marker="^")
    plt.scatter(beta, np.imag(np.sqrt(c * d + 0j)), color="r", marker="^")
plt.show()
```
