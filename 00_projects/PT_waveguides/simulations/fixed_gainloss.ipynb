```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
gamma = 10.0
k = 1.0
beta = 0.5
delta = 0.0
Ns = np.arange(0.1, 20, 0.01)
real_parts_a = []
imag_parts_a = []
real_parts_b = []
imag_parts_b = []
for i in range(len(Ns)):
    J11 = 0
    J12 = 2 * beta
    J13 = 0
    J21a = 0
    J21b = -4 * beta
    J22 = 0
    J23 = 2 * Ns[i] * np.sqrt(k ** 2 - beta ** 2)
    J31 = 0
    J32 = 4 * gamma / (Ns[i] + 2) ** 2 - 2 * np.sqrt(k ** 2 - beta ** 2) / Ns[i]
    J33 = 0
    # Jacobiano rama negativa
    Ja = np.array([
        [J11, J12, J13],
        [J21a, J22, J23],
        [J31, J32, J33]
    ])

    Jb = np.array([
        [J11, J12, J13],
        [J21b, J22, J23],
        [J31, J32, J33]
    ])

    # Autovalores
    eigs_a = np.linalg.eigvals(Ja)
    eigs_b = np.linalg.eigvals(Jb)

    # Guardamos partes real e imaginaria
    real_parts_a.append(np.real(eigs_a))
    imag_parts_a.append(np.imag(eigs_a))
    real_parts_b.append(np.real(eigs_b))
    imag_parts_b.append(np.imag(eigs_b))
# Convertimos a arrays
real_parts_a = np.array(real_parts_a)
imag_parts_a = np.array(imag_parts_a)
real_parts_b = np.array(real_parts_b)
imag_parts_b = np.array(imag_parts_b)

# Gráfico de partes reales
plt.figure(figsize=(10, 5))
plt.plot(Ns, real_parts_a[:, 0], label='Re(λ₁)')
plt.plot(Ns, real_parts_a[:, 1], label='Re(λ₂)')
plt.plot(Ns, real_parts_a[:, 2], label='Re(λ₃)')
plt.title('Parte real de los autovalores vs N (rama positiva)')
plt.xlabel('N')
plt.ylabel('Parte real')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Gráfico de partes imaginarias
plt.figure(figsize=(10, 5))
plt.plot(Ns, imag_parts_a[:, 0], label='Im(λ₁)')
plt.plot(Ns, imag_parts_a[:, 1], label='Im(λ₂)')
plt.plot(Ns, imag_parts_a[:, 2], label='Im(λ₃)')
plt.title('Parte imaginaria de los autovalores vs N (rama positiva)')
plt.xlabel('N')
plt.ylabel('Parte imaginaria')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Gráfico de partes reales
plt.figure(figsize=(10, 5))
plt.plot(Ns, real_parts_b[:, 0], label='Re(λ₁)')
plt.plot(Ns, real_parts_b[:, 1], label='Re(λ₂)')
plt.plot(Ns, real_parts_b[:, 2], label='Re(λ₃)')
plt.title('Parte real de los autovalores vs N (rama negativa)')
plt.xlabel('N')
plt.ylabel('Parte real')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Gráfico de partes imaginarias
plt.figure(figsize=(10, 5))
plt.plot(Ns, imag_parts_b[:, 0], label='Im(λ₁)')
plt.plot(Ns, imag_parts_b[:, 1], label='Im(λ₂)')
plt.plot(Ns, imag_parts_b[:, 2], label='Im(λ₃)')
plt.title('Parte imaginaria de los autovalores vs N (rama negativa)')
plt.xlabel('N')
plt.ylabel('Parte imaginaria')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
