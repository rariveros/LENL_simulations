```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
gamma = 10.0
k = 1.0
dN = 0.5
Ns = np.arange(0.5, 60, dN)
beta = 0.9
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for i in range(len(Ns)):
    if Ns[i] == Ns[0]:
        N = Ns[i]
    else:
        N = Ns[i - 1] + dN * (np.sqrt(k ** 2 - beta ** 2) / (2 * beta))
    a = 2 * beta
    c = 2 * N * np.sqrt(k ** 2 - beta ** 2)
    d = 4 * gamma / (N + 2) ** 2 - 2 * np.sqrt(k ** 2 - beta ** 2) / N
    if np.real(np.sqrt(c * d + 0j)) > 0.01:
        color_01 = "r"
    else:
        color_01 = "b"
    if np.real(np.sqrt(c * d - 2 * a ** 2 + 0j)) > 0.01:
        color_02 = "r"
    else:
        color_02 = "b"
    ax.scatter3D(N, 0, np.arcsin(beta / k), label='3D curve', color=color_01)
    ax.scatter3D(N, 0, np.pi - np.arcsin(beta / k), label='3D curve', color=color_02)
ax.set_xlabel('$N$')
ax.set_ylabel('$z$')
ax.set_zlabel(r'$\phi$')
plt.show()
plt.close()
```
